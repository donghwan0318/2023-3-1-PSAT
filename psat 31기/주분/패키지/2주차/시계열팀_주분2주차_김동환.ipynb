```python
#문제6
import torch
x_train = torch.tensor(x_train).reshape(200, 1, 5) #, dtype=torch.float32
x_test=torch.tensor(x_test).reshape(53,1,5)
y_train=torch.tensor(y_train)
y_test=torch.tensor(y_test)
```

```python
#shape 확인
print('x_train : ' ,x_train.shape)
print('y_train : ' ,y_train.shape)
print('x_test : ' ,x_test.shape)
print('y_test : ' ,y_test.shape)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
#문제7
import torch.nn as nn
class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size, output_size, num_layers):
        super(LSTMModel, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers)
        self.fc = nn.Linear(hidden_size, output_size)
        
    def forward(self, input):
        batch_size = input.size(1) # batch_size와 sequence length의 순서를 변경합니다.
        h0 = torch.zeros(1, batch_size, self.hidden_size).to(device) # hidden state 크기를 변경합니다.
        c0 = torch.zeros(1, batch_size, self.hidden_size).to(device) # hidden state 크기를 변경합니다.
        output, (hn, cn) = self.lstm(input, (h0, c0))
        output = self.fc(output[-1])
        return output
        
```

```python
#문제8, 문제9
input_size = 5
hidden_size = 2
output_size = 1
num_layers = 1
model = LSTMModel(input_size, hidden_size, output_size, num_layers).to(device)

# loss function, optimizer 설정
loss_fn = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.0001)

# 모델 학습
num_epochs = 5000
x_train = x_train.float().to(device)
y_train = y_train.float().to(device)
for epoch in range(num_epochs):
    optimizer.zero_grad()
    output = model(x_train)
    loss = loss_fn(output, y_train)
    loss.backward()
    optimizer.step()
    
    # 1000번마다 loss 출력
    if epoch % 1000 == 0:
        print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}")
```
